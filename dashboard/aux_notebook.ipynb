## Preprocessing

```python
chroma = pp.PCP('chord_x.wav')[:, :1216]
#padding = pp.add_padding(chroma)
```

```python
padded_chroma = pp.add_padding(tf.expand_dims(chroma, 0))
expand_chroma = tf.expand_dims(padded_chroma[0], 1)
reshaped_chroma = tf.reshape(expand_chroma, [1, 1216, 12])
#reshaped_chroma = chroma.reshape(1, 12, chroma.shape[1])
#reshaped_chroma.reshape(None, 12, chroma.shape[1] )
```
